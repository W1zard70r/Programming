## Нет особого смысла менять тип столбца $year$, так как он итак $int64$

```python
categorical_columns = ['region', 'manufacturer', 'model', 'fuel', 'title_status', 'transmission', 'state', 'price_category']

for col in categorical_columns:
    v_df[col]= v_df[col].astype('category')

v_df.dtypes
```

```python
v_df['year'].describe()
```

```python
Q1 = 2008
Q3 = 2017
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = v_df[(v_df['year'] < lower_bound) | (v_df['year'] > upper_bound)]

print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Нижняя граница: {lower_bound}, Верхняя граница: {upper_bound}")
print(f"Количество выбросов: {len(outliers)}")
```
